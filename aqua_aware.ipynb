AquaAware :- Household Water Usage Survey & Visualization Using python

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print("Hello User! Take a small survey")
name=input("Enter your name")
print("\nEnter the appropriate answer based on daily water usage.\n")
a=int(input("Enter your daily water consumption in litres for a single shower\n"))
b=int(input("Enter the amount of water you use in litres to wash the dishes once a day\n"))
c=int(input("Enter the amount of water you use in litres for brushing teeth once a day\n"))
d=int(input("Enter the amount of water you use in litres for washing clothes once a day\n"))
```

```python
household_tasks=['Shower','Dishwashing','Brushing teeth','Washing clothes']
print("\n")
predefined_values=np.array([60,20,5,40])
user_values=np.array([a,b,c,d])
X_axis=np.arange(len(household_tasks))
plt.bar(X_axis- 0.2,predefined_values,0.4,label='water usage as per safety guidelines')
plt.bar(X_axis + 0.2,user_values,0.4,label='water usage as per '+name)
plt.xticks(X_axis,household_tasks)
plt.xlabel("Household tasks")
plt.ylabel("litres of water")
plt.legend()
plt.show()
```

```python
count=0
def checking_safety(safety,usage,task):
    if safety>=usage:
        print("You are using water under safety level for "+task,"\n")
        count=0
    else:
        count=1
        exceeding_value=usage-safety
        print("You are exceeding safety limit by",exceeding_value,"litres for",task)
        print("Some Efficient methods which may help are:")
        if task=="showering":
            print("1.Install a Low-Flow showerhead\n2.Take shorter showers\n")
        elif task=="dishwashing":
            print("1.Use Dishwasher cycle instead of washing dishes by hand\n2.Close the running tap when not in use\n")
        elif task=="brushing teeth":
            print("1.Turn off the running tap\n2.Consider a Faucet censor\n")
        else:
            print("1.Pre-treat strains\n2.Choose a high-efficiency washer\n")
    return count
```

```python
g=checking_safety(60,a,"showering")
h=checking_safety(20,b,"dishwashing")
i=checking_safety(5,c,"brushing teeth")
j=checking_safety(40,d,"washing clothes")
if(g==1 or h==1 or i==1 or j==1):
    print("Dear",name+"!\n\033[1mWater is driving force of all Nature.\033[3m\nDo not waste it.\n")
else:
    print("Dear",name+"!\n\033[1mWe need citizens like you\033[0m\n")
```
